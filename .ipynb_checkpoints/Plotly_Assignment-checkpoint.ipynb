## Q1: Combine the two datsets as one dataframe

```python
movies=pd.concat([movies, credits],axis=1)
movies
```

## Q2: Remove the following columns:
- 'homepage','tagline','id','overview','status','original_title','movie_id'.

```python
movies.drop(['homepage','tagline','id','overview','status','original_title','movie_id'],axis=1,inplace=True)
```

```python
movies.head(3)
```

## Q3: Create Genres dataframe that hase two coulumns Genres and Frequency
- Hint: Use Counter to calculate Frequency
- Add new catagory to Genres called Others when the Frequency < 200

```python
genres = []
for movie in movies.genres:
    genreslist = ast.literal_eval(movie)
    for genre in genreslist:
        genres.append(genre['name'])

gendf = pd.DataFrame(data = genres,columns=['genre'])
gendf.groupby('genre')
gendf
```

```python
genfre = pd.DataFrame.from_dict(Counter(gendf.genre), orient='index').reset_index()
genfre = genfre.rename(columns={'index':'Genre', 0:'Frequency'})
genfre
```

```python
for i, row in genfre.iterrows():
    if row['Frequency'] < 200:
        genfre.loc[i, 'Genre'] = 'Others'
```

```python
genfre
```

## Q4: Use bar chart draws the relationship between genres of movies and Frequency.


```python
px.bar(data_frame=genfre,x='Genre', y='Frequency' ,title='Genres of Movies', labels='Genre', color='Genre')
```

## Q5: Use pie chart to draw top 5 languages.

```python
topl = pd.DataFrame.from_dict(Counter(movies.original_language), orient='index').reset_index()
topl = topl.rename(columns={'index':'language', 0:'Frequency'})
topl
```

```python
px.pie(topl.head(5), names='language',values='Frequency', title='The Top 5 of Languages')
```

## Q6: Use WordCloud to draw genres.

```python
x = gendf["genre"].value_counts()
text = ' '.join(gendf['genre'])

wordcloud = WordCloud(
    background_color='white',  
    width=800, height=400,      
    max_words=100,              
    contour_width=3,            
    contour_color='steelblue'  
).generate_from_frequencies(x)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show()
```
