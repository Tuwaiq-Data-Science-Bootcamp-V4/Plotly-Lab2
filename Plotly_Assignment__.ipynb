## Q3: Create Genres dataframe that hase two coulumns Genres and Frequency
- Hint: Use Counter to calculate Frequency
- Add new catagory to Genres called Others when the Frequency < 200

```python
# The ast.literal_eval method can evaluate Python abstract syntax strings grammar 

GenresList=[]

#iterrows() Iterate over DataFrame rows
for i,r in movies.iterrows():
    genres = ast.literal_eval(r["genres"])
    for g in genres :
        GenresList.append(g["name"])
        
#calculate Frequency
df = pd.DataFrame({'Genres': Counter(GenresList).keys(), 'Frequency': Counter(GenresList).values()})
df.loc[df['Frequency'] < 200, 'Genres'] = 'Others'
df
```

## Q4: Use bar chart draws the relationship between genres of movies and Frequency.


```python
px.bar(df, x='Genres', y='Frequency',color='Genres',title='Genres of Movies')
```
